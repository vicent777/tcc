```python
import pandas as pd
import re
import dateparser
```

```python
data = pd.read_excel('Consolidado_Hoteis.xlsx')
```

```python
data.head(20)
```

```python
data.info()
```

```python
# Regex para identificar datas no formato textual em português
padrao_data = r'^\d{1,2} de [a-zç]+ de \d{4}$'

# Padronizar tudo para minúsculas e remover espaços
data['data_avaliacao'] = data['data_avaliacao'].astype(str).str.strip().str.lower()

# Aplicar a regex para filtrar linhas que "parecem" uma data
mascara_datas_validas = data['data_avaliacao'].str.match(padrao_data)

# Filtrar apenas essas linhas
data_validas = data[mascara_datas_validas].copy()

# Usar dateparser para converter para datetime (reconhece português automaticamente)
data_validas['data_avaliacao'] = data_validas['data_avaliacao'].apply(
    lambda x: dateparser.parse(x, languages=['pt'])
)

# Opcional: formatar como string, no mesmo padrão original
data_validas['data_avaliacao'] = data_validas['data_avaliacao'].dt.strftime('%d-%m-%Y')

# Substituir o DataFrame original pelos dados limpos
data = data_validas

print(f"✅ {len(data)} linhas com datas válidas mantidas com sucesso (com parsing em português).")
```

```python
data['data_avaliacao'] = pd.to_datetime(data['data_avaliacao'], format='%d-%m-%Y', errors='coerce')
data.info()
```
